```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import plot_tree
import pydotplus
from sklearn import tree
from IPython.display import Image
```

```python
iris = datasets.load_iris()
features = iris['data']
target = iris['target']
```

```python
# Inisialisasi model Decision Tree dengan berbagai parameter
decisiontree = DecisionTreeClassifier(
    random_state=0,
    max_depth=None,                # Kedalaman maksimal dari pohon (None berarti tidak terbatas)
    min_samples_split=2,           # Jumlah minimum sampel untuk memisahkan node
    min_samples_leaf=1,            # Jumlah minimum sampel dalam satu daun
    min_weight_fraction_leaf=0,    # Fraksi minimum dari berat total sampel yang diperlukan untuk setiap daun
    max_leaf_nodes=None,           # Jumlah maksimum daun (None berarti tidak terbatas)
    min_impurity_decrease=0        # Penurunan impurity minimum sebelum dilakukan split
)
```

```python
model = decisiontree.fit(features, target)
```

```python
observation = [[5, 4, 3, 2]]
model.predict(observation)
model.predict_proba(observation)
```

```python
# Membuat visualisasi menggunakan pydotplus dan Graphviz
dot_data = tree.export_graphviz(
    decisiontree,
    out_file=None,
    feature_names=iris['feature_names'],
    class_names=iris['target_names'],
    filled=True,
    rounded=True,
    special_characters=True
)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
# Simpan visualisasi pohon keputusan sebagai file PNG
graph.write_png('iris_tree_visualization.png')
```

```python
# Visualisasi menggunakan plot_tree
plt.figure(figsize=(12, 8))
plot_tree(
    decisiontree,
    feature_names=iris['feature_names'],
    class_names=iris['target_names'],
    filled=True,
    rounded=True,
    fontsize=10
)
plt.show()
```
